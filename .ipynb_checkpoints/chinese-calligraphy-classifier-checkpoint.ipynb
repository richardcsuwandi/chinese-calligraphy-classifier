```python
# Import the libraries
from fastai.vision import *
import os
```

```python
# Get the working directory path
path = os.getcwd()
path
```

```python
np.random.seed(42)

data = ImageDataBunch.from_folder(
    path, 
    valid_pct=0.2,
    ds_tfms=get_transforms(do_flip=False),
    size=224,
    num_workers=4    
)

data.normalize(imagenet_stats)
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
data.show_batch(rows=3, figsize=(9,10))
```

## Training the Model

```python
learn = cnn_learner(data, models.resnet50, metrics=accuracy)
```

```python
learn.fit_one_cycle(3)
```

```python
# Save the model
learn.save("2020-07-30 version-1")
```

## Tuning the Model

```python
# Load the saved model
learn = cnn_learner(data, models.resnet50, metrics=accuracy).load("2020-07-30 version-1")
```

```python
learn.unfreeze()
learn.lr_find()
```

```python
learn.recorder.plot(suggestion=True)
```

```python
min_grad_lr = learn.recorder.min_grad_lr
min_grad_lr
```

```python
learn.fit_one_cycle(2, min_grad_lr)
```

```python
learn.save("2020-07-30 version-2")
```

## Cleaning the Data

```python
from fastai.widgets import *
```

```python
ds, idxs = DatasetFormatter().from_toplosses(learn)
ImageCleaner(ds, idxs, path)
```

```python
df = pd.read_csv(path + "/cleaned.csv", header="infer")
df.head()
```

```python
np.random.seed(42)
db = (ImageList.from_df(df, path)
                   .random_split_by_pct(0.2)
                   .label_from_df()
                   .transform(get_transforms(), size=224)
                   .databunch(bs=9).normalize(imagenet_stats)
```

```python
db.classes, db.c, len(db.train_ds), len(db.valid_ds)
```

```python
db.show_batch()
```

## Retraining the Model

```python
learn = cnn_learner(data, models.resnet50, metrics=accuracy).load("2020-07-30 version-2")
```

```python
learn.data = db
```

```python
learn.freeze()
learn.fit_one_cycle(4)
```

```python
learn.unfreeze()
learn.lr_find()
```

```python
learn.recorder.plot(suggestion=True)
```

```python
min_grad_lr = learn.recorder.min_grad_lr
min_grad_lr
```

```python
learn.fit_one_cycle(2, min_grad_lr)
```

## Interpretation

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
losses, idxs = interp.top_losses()
interp.plot_top_losses(9)
```

```python
learn.show_results(ds_type=DatasetType.Train, rows=4, figsize=(8,10))
```
